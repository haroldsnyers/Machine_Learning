```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter


class Plot:

    def __init__(self, file_path):
        self.data = pd.read_csv(file_path, header = None, sep=" ")
        
    def plot(self):
        plt.plot(self.data[0], self.data[1], 'bo')
        plt.ylabel("x1")
        plt.xlabel("x2")
        plt.show()
        
```

```python
k_means = Plot("data_kmeans.txt")
pca = Plot("data_pca.txt")
test = Plot("iris.data")

k_means.plot()
pca.plot()


```
